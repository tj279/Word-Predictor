```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
df=pd.read_csv("HappyDB\happydb\data\cleaned_hm.csv"
               
)
df=df['cleaned_hm'].iloc[:10000]
df = df.dropna()
tokenizer=Tokenizer()
tokenizer.fit_on_texts(df.to_list())
word_index=tokenizer.word_index
print(len(word_index))
```

```python
import os
print(os.getcwd())
```

```python
len(df)
```

```python
inputs=[]
for i in range(0, len(df)):
    sequence=tokenizer.texts_to_sequences([df[i]])[0]
    print(sequence)
    for j in range(1, len(sequence)):
        inputs.append(sequence[:j+1])
```

```python
inputs
```

```python
lenth=0
for s in inputs:
    lenth=max(len(s), lenth)
    if len(s)==1204:
        print(s)
```

```python
print(lenth)
```

```python
inputs
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
inputs=pad_sequences(inputs, maxlen=lenth, padding='pre')
```

```python
outputs=inputs[:, -1:]   
```

```python
inputs=inputs[:, :-1]
inputs
```

```python
from tensorflow.keras.utils import to_categorical
outputs=to_categorical(outputs, num_classes=len(word_index))
```
